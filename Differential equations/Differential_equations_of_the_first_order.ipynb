# Диференціальні рівняння з роздільними змінними

Диференціальні рівняння (ДР). Ці два слова зазвичай викликають страх у середнього громадянина. Диференціальні рівняння здаються чимось надзвичайним і важким для освоєння, і багатьом студентам. Уууу... диференціальні рівняння, як мені усе це пережити?!

Така думка і такий настрій коренево невірні, оскільки насправді **ДИФЕРЕНЦІАЛЬНІ РІВНЯННЯ - ЦЕ ПРОСТО І НАВІТЬ ЗАХОПЛЮЮЧЕ**. Що потрібно знати і вміти, щоб навчитися розв'язувати диференціальні рівняння? Для успішного вивчення дифурів ви повинні добре вміти інтегрувати і диференціювати. Чим якісніше вивчені теми Похідна функції однієї змінної та Невизначений інтеграл, тим легше розібратися в диференціальних рівняннях. Скажу більше, якщо у вас досить гарні навички інтегрування, то тему практично освоєно! Чим більше різних типів інтегралів ви вмієте розв'язувати - тим краще. Чому? Доведеться багато інтегрувати. І диференціювати. Також настійно рекомендую навчитися знаходити похідну від функції, заданої неявно.

У 95% випадків на контрольних роботах зустрічаються 3 типи диференціальних рівнянь першого порядку: рівняння з роздільними змінними, які ми розглянемо на цьому уроці; однорідні рівняння і лінійні неоднорідні рівняння. Початківцям у вивченні дифурів раджу ознайомитися з уроками саме в такій послідовності, причому після вивчення перших двох статей не забракне закріпити свої навички на додатковому практикумі - рівняннях, які зводяться до однорідних.

Існують ще більш рідкісні типи диференціальних рівнянь: рівняння в повних диференціалах, рівняння Бернуллі та деякі інші. Найбільш важливими з останніх двох видів є рівняння в повних диференціалах, оскільки окрім даного ДР я розглядаю новий матеріал - часткове інтегрування.

Отже, орієнтаційні вказівки встановлені - поїхали:
Давайте розглянемо «звичайні» рівняння спочатку. Вони містять змінні та числа. Найпростіший приклад: $3x = 12$. Що означає розв'язати таке рівняння? Це означає знайти множину всіх чисел, які задовольняють даному рівнянню. Легко побачити, що дитяче рівняння $3x = 12$ має єдиний корінь $x = 4$. Проведемо перевірку, підставивши чотири в рівняння:
$3 \cdot 4 = 12$

$12 = 12$ - отримано правильну рівність, отже, рішення знайдено правильно.

Диференціальні рівняння будуть розглядатися схожим чином!

**Диференціальне рівняння першого порядку** в загальному випадку **містить**: </br>
1) незалежну змінну $x$;</br>
2) залежну змінну $y$ (функцію);</br>
3) першу похідну функції: $y'$.

У деяких рівняннях першого порядку може бути відсутнім «x» або (і) «y», але це не суттєво - головне, щоб у ДР була перша похідна $y'$ і не було похідних вищих порядків - $y''$, $y'''$ і так далі.

**Що означає розв'язати диференціальне рівняння?** *Розв'язати диференціальне рівняння - це означає знайти множину всіх функцій, які задовольняють даному рівнянню*(порою, досить однієї). Тобто коренями диференціального рівняння є функції. Для ДР першого порядку така множина функцій часто має вигляд $y = f(x, C)$, який називають загальним рішенням диференціального рівняння (де "С" може приймати різні дійсні значення).

**Приклад 1:**

Розв'язати диференціальне рівняння $x \cdot y' = y$. Повний арсенал. З чого почати розв'язок? По-перше, потрібно переписати похідну трохи іншим способом. Пригадуємо громіздку нотацію $y' = \frac{dy}{dx}$, яка, можливо, здавалася багатьом з вас дивною та непотрібною. У дифурі це саме те, що рулить!

Отже:
$$
x \cdot \frac{dy}{dx} = y
$$
