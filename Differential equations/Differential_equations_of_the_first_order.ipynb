# Диференціальні рівняння з роздільними змінними

Диференціальні рівняння (ДР). Ці два слова зазвичай викликають страх у середнього громадянина. Диференціальні рівняння здаються чимось надзвичайним і важким для освоєння, і багатьом студентам. Уууу... диференціальні рівняння, як мені усе це пережити?!

Така думка і такий настрій коренево невірні, оскільки насправді **ДИФЕРЕНЦІАЛЬНІ РІВНЯННЯ - ЦЕ ПРОСТО І НАВІТЬ ЗАХОПЛЮЮЧЕ**. Що потрібно знати і вміти, щоб навчитися розв'язувати диференціальні рівняння? Для успішного вивчення дифурів ви повинні добре вміти інтегрувати і диференціювати. Чим якісніше вивчені теми Похідна функції однієї змінної та Невизначений інтеграл, тим легше розібратися в диференціальних рівняннях. Скажу більше, якщо у вас досить гарні навички інтегрування, то тему практично освоєно! Чим більше різних типів інтегралів ви вмієте розв'язувати - тим краще. Чому? Доведеться багато інтегрувати. І диференціювати. Також настійно рекомендую навчитися знаходити похідну від функції, заданої неявно.

У 95% випадків на контрольних роботах зустрічаються 3 типи диференціальних рівнянь першого порядку: рівняння з роздільними змінними, які ми розглянемо на цьому уроці; однорідні рівняння і лінійні неоднорідні рівняння. Початківцям у вивченні дифурів раджу ознайомитися з уроками саме в такій послідовності, причому після вивчення перших двох статей не забракне закріпити свої навички на додатковому практикумі - рівняннях, які зводяться до однорідних.

Існують ще більш рідкісні типи диференціальних рівнянь: рівняння в повних диференціалах, рівняння Бернуллі та деякі інші. Найбільш важливими з останніх двох видів є рівняння в повних диференціалах, оскільки окрім даного ДР я розглядаю новий матеріал - часткове інтегрування.

Отже, орієнтаційні вказівки встановлені - поїхали:
Давайте розглянемо «звичайні» рівняння спочатку. Вони містять змінні та числа. Найпростіший приклад: $3x = 12$. Що означає розв'язати таке рівняння? Це означає знайти множину всіх чисел, які задовольняють даному рівнянню. Легко побачити, що дитяче рівняння $3x = 12$ має єдиний корінь $x = 4$. Проведемо перевірку, підставивши чотири в рівняння:
$3 \cdot 4 = 12$

$12 = 12$ - отримано правильну рівність, отже, рішення знайдено правильно.

Диференціальні рівняння будуть розглядатися схожим чином!

**Диференціальне рівняння першого порядку** в загальному випадку **містить**: </br>
1) незалежну змінну $x$;</br>
2) залежну змінну $y$ (функцію);</br>
3) першу похідну функції: $y'$.

У деяких рівняннях першого порядку може бути відсутнім «x» або (і) «y», але це не суттєво - головне, щоб у ДР була перша похідна $y'$ і не було похідних вищих порядків - $y''$, $y'''$ і так далі.

**Що означає розв'язати диференціальне рівняння?** *Розв'язати диференціальне рівняння - це означає знайти множину всіх функцій, які задовольняють даному рівнянню*(порою, досить однієї). Тобто коренями диференціального рівняння є функції. Для ДР першого порядку така множина функцій часто має вигляд $y = f(x, C)$, який називають загальним рішенням диференціального рівняння (де "С" може приймати різні дійсні значення).

**Приклад 1:**

Розв'язати диференціальне рівняння $x \cdot y' = y$. Повний арсенал. З чого почати розв'язок? По-перше, потрібно переписати похідну трохи іншим способом. Пригадуємо громіздку нотацію $y' = \frac{dy}{dx}$, яка, можливо, здавалася багатьом з вас дивною та непотрібною. У дифурі це саме те, що рулить!

Отже:
$$
x \cdot \frac{dy}{dx} = y
$$

На другому кроці ми перевіряємо, чи можливо розділити змінні. Що означає розділити змінні? Грубо кажучи, в лівій частині ми маємо залишити лише "ігреки", а в правій частині організувати лише "ікси". Розділення змінних виконується за допомогою "шкільних" маніпуляцій: витягнення за дужки, перенесення доданків з однієї частини до іншої зі зміною знака, переміщення множників з однієї частини до іншої за правилом пропорційності і так далі.

Диференціали $dy$ і $dx" це повноправні множники та активні учасники бойових дій. У розглянутому прикладі змінні легко розділяються шляхом перекидання множників за правилом пропорції:

$$
\frac{dy}{y} = \frac{dx}{x}
$$

Змінні розділені. В лівій частині - лише "ігреки", в правій частині - лише "ікси".

Наступний етап - інтегрування диференціального рівняння. Все просто, навішуємо інтеграли на обидві частини:

$$
\int \frac{dy}{y} = \int \frac{dx}{x}
$$

Звісно ж, інтеграли потрібно взяти. У даному випадку вони стандартні:

$$
\ln |y| = \ln |x| + C
$$

Як ми пам'ятаємо, до будь-якої первісної додається константа. Тут два інтеграли, але константу C досить записати один раз (оскільки константа + константа все одно дорівнює іншій константі). У більшості випадків її поміщають в праву частину.

Строго кажучи, після того, як взяті інтеграли, диференціальне рівняння вважається вирішеним. Єдине, у нас "ігрек" не виражений через "ікс", отже, рішення представлено в неявному вигляді. Рішення диференціального рівняння в неявному вигляді називається загальним інтегралом диференціального рівняння. Тобто $\ln |y| = \ln |x| + C$ - це загальний інтеграл.

Відповідь у такій формі цілком прийнятна, але немає чиєїсь іншої варіації? Давайте спробуємо отримати загальне рішення.

Будь ласка, запам'ятайте перший технічний прийом, він дуже поширений і часто застосовується в практичних завданнях: якщо в правій частині після інтегрування з'являється логарифм, то константу в багатьох випадках (але далеко не завжди!) доцільно записати також під логарифмом. І записати ОБОВ'ЯЗКОВО, якщо отрималися лише логарифми (як у розглянутому прикладі).

Отже, ЗАМІСТЬ запису $\ln |y| = \ln |x| + C$ зазвичай пишуть $\ln |y| = \ln |Cx|$ (і це коректно, оскільки $\ln |C|$ з таким самим успіхом приймає всі дійсні значення, що і C).

Для чого це потрібно? Для того, щоб було легше виразити "ігрек". Використовуємо властивість логарифмів $\ln |a| + \ln |\beta| = \ln |a\beta|$. В даному випадку:

$$
\ln |y| = \ln |Cx|
$$

Тепер логарифми і модулі можна прибрати:

$$
y = Cx
$$

Функція представлена в явному вигляді, і це є загальним розв'язком. Відповідь: загальний розв'язок - $y = Cx$, де $C$ - довільна константа.

Відповіді на багато диференціальних рівнянь можна досить легко перевірити. У нашому випадку це робиться дуже просто: беремо знайдений розв'язок $y = Cx$ і обчислюємо його похідну:

$$
y' = C
$$

Потім підставляємо $y = Cx$ і похідну $y' = C$ в початкове рівняння $xy' = y$:

$$
x \cdot C = Cx
$$

$Cx = Cx$ - отримано правильну рівність, отже, множина функцій $y = Cx$ задовольняє диференціальному рівнянню $xy' = y$, що було потрібно перевірити.

Давши константі $C$ різні значення, можна отримати безмежно багато часткових розв'язків диференціального рівняння. Очевидно, що будь-яка з функцій $y = x, y = -3x, y = \frac{x}{5}$ і так далі задовольняє диференціальному рівнянню $xy' = y$.

Іноді загальний розв'язок називають сім'єю функцій. У цьому прикладі загальний розв'язок $y = Cx$, де $C = const$, - це сім'я лінійних функцій, а точніше, сім'я прямих пропорційностей.

Після ретельного розгляду першого прикладу можна відповісти на декілька наївних запитань щодо диференціальних рівнянь:

1) У цьому прикладі нам вдалося розділити змінні. Чи завжди це можливо зробити? Ні, не завжди. І частіше, ніж не можна. Наприклад, в однорідних рівняннях першого порядку спочатку потрібно зробити заміну. В інших типах рівнянь, наприклад, в лінійних неоднорідних рівняннях першого порядку, потрібно використовувати різні прийоми і методи для знаходження загального розв'язку. Рівняння з роздільними змінними, які ми розглядаємо на першому уроці, - це найпростіший тип диференціальних рівнянь.

2) Чи завжди можна інтегрувати диференціальне рівняння? Ні, не завжди. Дуже легко придумати "складне" рівняння, яке не можна інтегрувати. Крім того, існують невбираємі інтеграли. Але подібні диференціальні рівняння можна розв'язати приблизно за допомогою спеціальних методів. Даламбер та Коші гарантують... ...фу, lurkmore.to нещодавно прочитав, майже не додав "з іншого світу".

3) В данному прикладі ми отримали розв'язок у вигляді загального інтеграла $\ln|y|=\ln|x|+\ln|C|$. Чи завжди можна з загального інтеграла знайти загальний розв'язок, тобто виразити «игрек» в явному вигляді? Ні, не завжди. Наприклад: $y+\ln|y|=\operatorname{arcsin}(x)+xy^2 + C$. А як тут виразити «игрек»?! У таких випадках відповідь слід записати у вигляді загального інтеграла. Крім того, іноді загальний розв'язок можна знайти, але він записується настільки громіздко і коряво, що краще залишити відповідь у вигляді загального інтеграла.

4) ...Можливо, цього наразі досить. У першому прикладі зустрівся ще один важливий момент, але щоб не пригнічувати "початківців" лавиною нової інформації, я залишу його до наступного уроку.

**Приклад 2.** Знайдіть частковий розв'язок диференціального рівняння $y^{\prime}=-2y$, який задовольняє початковій умові $y(0)=2$.

Розв'язок: За умови потрібно знайти частковий розв'язок диференціального рівняння, який відповідає заданому початковому умові. Така постановка задачі називається задачею Коші.

Спочатку знаходимо загальний розв'язок. У рівнянні відсутня змінна $x$, але це не має змушувати нас смішувати. Головне, в ньому є перша похідна.

Перепишемо похідну у відповідному вигляді:

$\frac{dy}{dx}=-2y$

Очевидно, що змінні можна розділити: хлопці - ліворуч, дівчата - праворуч!

$\frac{dy}{y}=-2dx$

Інтегруємо рівняння:

$\int\frac{dy}{y}=-2\int dx$

$\ln|y|=-2x+C^{\star}$

Загальний розв'язок отримано. Тут я позначив константу з верхньою зірочкою, оскільки дуже скоро вона перетвориться в іншу константу.

Тепер спробуємо перетворити загальний інтеграл в загальний розв'язок (виразити "y" у явному вигляді). Пам'ятаємо шкільне правило: $\ln a=b\Longrightarrow a=e^{b}$. У цьому випадку:
$$|y|=e^{-2x+C^{\star}}$$

Ступінь в показнику кажеться дещо непристойним, тому її зазвичай припускають до землі. Детально це виглядає так. Використовуючи властивість степенів, перепишемо функцію наступним чином:

$$|y|=e^{C^{\star\star}}\cdot e^{-2x}$$

Якщо $C^{\star}$ - це константа, то $ e^{C^\star}>0 $ - також позитивна константа, позначимо її як $C^{\star\star}$:

$|y|=C^{\star}\cdot e^{-2\pi}$

Потім розкриваємо модуль:

$y=\pm C^{\star\star}\cdot e^{-2x}$ і знову перевизначаємо константу $\pm C^{\star\star}=C$, припускаючи, що "це" може мати як позитивні, так і від'ємні значення:

$y=Ce^{-2x}$

Запам'ятайте, "зности" константу - це другий технічний прийом, який часто використовується під час розв'язання диференціальних рівнянь. Зазвичай на чистовику відразу переходять від $\ln|y|=-2x+C^\star$ до $y=Ce^{-2x}$, але завжди готові пояснити цей перехід. Так само, як ви попросили мене пояснити, і я пояснив :)

Отже, загальний розв'язок: $y=Ce^{-2x}$, де C=const. Така гарна родина експоненціальних функцій.

На завершальному етапі потрібно знайти частковий розв'язок, який відповідає заданій початковій умові $y(0)=2$. Це також просто.

В чому полягає задача? Необхідно підібрати таке значення константи $C$, щоб виконувалася умова $y(0)=2$.

Це можна оформити по-різному, але, ймовірно, найбільше зрозуміло буде так. У загальний розв'язок підставляємо нуль замість "x" і двійку замість "y":
$$2=Ce^{-2 \cdot 0}$$
$$2=Ce^{ 0}$$
$$2=C \cdot 1$$
отже, $C=2$.

Стандартний варіант формулювання:
$$y(0)=Ce^{-2 \cdot 0}=Ce^{ 0}=C=2$$

Тепер в загальний розв'язок $y=Ce^{-2x}$ підставляємо знайдене значення константи $C=2$: $y=2e^{-2x}$
 - це і є потрібний нам частковий розв'язок.

**Відповідь:** частковий розв'язок: $y=2e^{-2x}$

Проведемо перевірку. Перевірка часткового розв'язку включає два етапи:

Спочатку потрібно перевірити, чи дійсно знайдена функція $y=2e^{-2x}$ задовольняє початковій умові $y(0)=2$? Замість "x" підставляємо нуль і дивимося, що вийде: $y(0)=2e^{-2 \cdot 0}=2e^0=2$
 - так, дійсно отримали двійку, отже, початкова умова виконана.

Другий етап вже відомий. Беремо знайдену функцію $y=2e^{-2x}$ і знаходимо похідну:

$y'=(2e^{-2x})'=2(e^{-2x})' =2e^{-2x}\cdot (-2x)' =-4e^{-2x}$

Підставляємо $y=2e^{-2x}$ і $y'=-4e^{-2x}$ в початкове рівняння $y'=-2y$:
$$-4e^{-2x}=-2 \cdot 2e^{-2x}$$
$-4e^{-2x}=-4e^{-2x}$ - отримали правильну рівність.

Висновок: частковий розв'язок знайдено вірно.

**Приклад 3**

Розв'язати диференціальне рівняння $y' + (2y + 1) \operatorname{ctg} x = 0$.

Розв'язання:

1. Переписуємо похідну в потрібному нам вигляді:

$$
\frac{dy}{dx} + (2y + 1) \operatorname{ctg} x = 0
$$

2. Оцінюємо, чи можливо розділити змінні. В даному випадку це можливо. Переносячи другий доданок в праву частину зі зміною знаку, ми отримуємо:

$$
\frac{dy}{dx} = -(2y + 1) \operatorname{ctg} x
$$

І перекидуємо множники за правилом пропорції:

$$
\frac{dy}{2y + 1} = -\operatorname{ctg} x dx
$$

3. Змінні розділені, інтегруємо обидві частини:

$$
\int \frac{dy}{2y + 1} = -\int \operatorname{ctg} x dx
$$

4. Знайдемо інтеграл лівої частини, використовуючи метод підведення функції під знак диференціала, і розберемось з інтегралом від котангенса за стандартним методом, який ми розглядали на уроці інтегрування тригонометричних функцій:

$$
\begin{aligned}
& \frac{1}{2} \int \frac{d(2y + 1)}{2y + 1} = -\int \frac{\cos x}{\sin x} dx \\
& \frac{1}{2} \ln |2y + 1| = -\ln |\sin x| + \ln |C^*|
\end{aligned}
$$

5. Результатом є одні логарифми. Згідно з моєю першою технічною рекомендацією, якщо в правій частині після інтегрування з'являється логарифм, то в багатьох випадках (але далеко не завжди!) доцільно записувати константу також під логарифмом.

$$
\ln \sqrt{|2y + 1|} = \ln \frac{1}{|\sin x|} + \ln |C^*|
$$

6. Завершуємо упаковку, використовуючи властивості логарифмів:

$$
\ln \sqrt{|2y + 1|} = \ln \left| \frac{C^*}{\sin x} \right|
$$

7. Підносячи обидві сторони до квадрата та переозначаючи константу, ми отримуємо:

$$
|2y + 1| = \left| \frac{C^*}{\sin x} \right|^2
$$

8. Тепер можемо записати загальний інтеграл у вигляді $F(x, y) = C$, де це можливо:

$$
(2y + 1) \sin^2 x = C
$$

Отже, загальний інтеграл цього диференціального рівняння має вигляд $(2y + 1) \sin^2 x = C$, де $C$ - довільна константа.

! Зауваження: загальний інтеграл часто можна записати не єдиним способом. Отже, якщо ваш результат не збігається з наперед відомою відповіддю, це ще не означає, що ви неправильно розв'язали рівняння.

Чи можна виразити "ігрик"? Так. Давайте виразимо загальний розв'язок:

$$
(2y+1)\cdot\sin^{2}x=C\quad\Longrightarrow\quad2y+1=\frac{C}{\sin^{2}x}\quad\Longrightarrow\quad2y=\frac{C}{\sin^{2}x}-1\quad\Longrightarrow\quad y=\frac{C}{2\sin^{2}x}-\frac{1}{2}
$$
